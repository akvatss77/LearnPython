# histograms

```python
import matplotlib.pyplot as plt
```

```python
a = [1,2,3,4,1,3,4,2,1,1]
```

```python
plt.hist(a)
plt.show()
```

**lets look it on some bigger example**

```python
b = [1,2,1,3,5,4,5,5,4,3,3,10,7,6,7,8,8,20,22,11,14]
```

```python
plt.hist(b)
```

**here if we can see above then 1 appears only 2 times still it has been clubbed with 2 and freq is 6**

```python
plt.hist(b,edgecolor = 'black')
```

**xticks feature to scale xaxis**

```python
plt.hist(b,edgecolor = 'black')
plt.xticks([1,2,3,4,5])
plt.show()
```

```python
import numpy as np
```

```python
xt = np.arange(23)
xt
```

```python
plt.hist(b,edgecolor = 'black')
plt.xticks(xt)
plt.show()
```

## by default there is 10 bins are shown whose size is calculated by max - min/10

**still we can define the number of bins we want**

```python
plt.hist(b,bins = 25,edgecolor = 'black')
plt.xticks(xt)
plt.show()
```

## if a bin is [a,b] then a is inclusive and b is exclusive except for the last bin where both are inclusive

**we can also decide the range of each bin by following way**

```python
plt.hist(b,bins = [1,3,6,10,15,20],edgecolor = 'black')
plt.xticks(xt)
plt.show()
```

## giving xmin xmax and ymin and ymax values

```python
plt.hist(b,bins = [1,3,6,10,15,20],edgecolor = 'black')
plt.axis([0,25,0,10])
plt.show()
```

# Bar graphs

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
```

```python
plt.bar(year,population)
plt.show()
```


## Width of bar graph

```python
plt.bar(year,population,width=1)
plt.show()
```

```python
year = ["year -2012","year -2013","year -2014","year -2015","year -2016","year -2017"]
plt.bar(year,population,width=0.5)
plt.show()
```

## So you can see that the labels are overlapping

```python
plt.bar(year,population,width=0.5)
plt.xticks(rotation = 40)
plt.show()
```
