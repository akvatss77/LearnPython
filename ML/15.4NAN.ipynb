##  handling NAN entries

**1.) Drop NA**<br>
**2.) fill NA**

```python
import pandas as pd
import numpy as np
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data")
df = iris.copy()
```

```python
df.iloc[2:4,1:3] = np.nan
df.head()
```

```python
df.describe()
```

**Dropping rows containg NaN using function dropna()**

```python
df.dropna(inplace=True)
df
```

```python
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df=iris.copy()
df.columns = ['sl','sw','pl','pw','sp']
df.iloc[2:4,1:3] = np.nan
df.head()
```

**Now we will try to replace these NaN rather than deleting them,generally we fill by the mean using fillna() function**

```python
df.sw.fillna(df.sw.mean(),inplace = True)
df.head()
```

```python
df['pl'].fillna(df['pl'].mean(),inplace = True)
df.head()
```
