```python
import matplotlib.pyplot as plt
import numpy as np
```

# Bubble chart : this is used to basically show two outputs

```python
year = [2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
```

```python
plt.scatter(year,salary,s = 100)
plt.show()
```

```python
plt.scatter(year,salary,s =population)
plt.show()
```

```python
plt.scatter(year,salary,s =population,c = 'r')
plt.show()
```

```python
plt.scatter(year,salary,s =population,c = ['b','r','g','y','b','r'])
plt.show()
```

```python
plt.scatter(year,salary,s =population,c = ['b','r','g','y','b','r'],alpha=0.3) ##alpha defined opaqueness of the bubbles
plt.show()
```

```python
c = np.arange(len(year))
plt.scatter(year,salary,s =population,c = c) 
plt.show()
```

## edgecolor()

```python
c = np.arange(len(year))
plt.scatter(year,salary,s =population,c = c,edgecolor='black') 
plt.show()
```

```python
c = np.arange(len(year))
plt.scatter(year,salary,s =population,marker = '^',c = c,edgecolor='black') 
plt.show()
```

```python
c = np.arange(len(year))
plt.scatter(year,salary,s =population,marker = '^',c = c,edgecolor='black') 
plt.text(year[0],salary[0],population[0])
plt.show()
```

```python
c = np.arange(len(year))
plt.scatter(year,salary,s =population,marker = '^',c = c,edgecolor='black') 
plt.text(year[0] + 0.1,salary[0] + 0.1,population[0]) #we can do this in a loop
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]

employees = np.array(employees)
year = np.array(year)
revenue = np.array(revenue)

plt.scatter(year,revenue,s = employees)

for i in range(1,len(year) - 1):
    if (revenue[i] > revenue[i-1]) and (revenue[i] > revenue[i+1]):
        print(year[i],revenue[i],employees[i])
        
plt.show()
```
