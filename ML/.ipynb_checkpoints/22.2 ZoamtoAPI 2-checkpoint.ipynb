```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import csv

#Reading the dataset and taking data from India and replacing few ncr places to make it delhi-NCR
df1 = pd.read_csv('Dataset/zomato.csv',encoding='latin-1')
df1 = df1[df1['Country Code'] == 1]
df1['City'].replace("New Delhi","Delhi-NCR",inplace = True)
df1['City'].replace("Gurgaon","Delhi-NCR",inplace = True)
df1['City'].replace("Faridabad","Delhi-NCR",inplace = True)
df1['City'].replace("Ghaziabad","Delhi-NCR",inplace = True)
df1['City'].replace("Noida","Delhi-NCR",inplace = True)

#taking city column in np array to update non delhi cities to rest of india
city_np = np.array(df1['City'])
index = city_np != 'Delhi-NCR'
city_np[index] = 'Rest of India'

#deleted the old city column and replaced with new city column
del(df1['City'])
df1['City'] = city_np

#Taking the Cuisines and CIty list out to seggregate based on City
arr = np.array(df1['Cuisines'])
arr_loc = np.array(df1['City'])

cusine_roi = []
cuisine_delhi = []

#function to add string to array after splitting it based on comma
def append_to_arr(arr,item):
    n_arr = item.split(',')
    for k in n_arr:
        arr.append(k.strip())
    return arr

#looping over on all the cuisines and city to seggregate
for i in range(len(arr)):
    if arr_loc[i] == 'Rest of India':
        cusine_roi = append_to_arr(cusine_roi,arr[i])
    else:
        cuisine_delhi = append_to_arr(cuisine_delhi,arr[i])
        
#Taking the counts of each of the cuisine for delhi and rest of india
cusine_in_delhi,counts_delhi= np.unique(cuisine_delhi,return_counts=True)
cuisne_in_roi,counts_roi= np.unique(cusine_roi,return_counts=True)

#taking top 10 cusines based on count for delhi and rest of india
index_delhi = counts_delhi.argsort()
index_delhi = index_delhi[::-1][:10]
max_served_delhi = []
max_served_delhi = cusine_in_delhi[index_delhi]

index_roi = counts_roi.argsort()
index_roi = index_roi[::-1][:10]
max_served_roi = []
max_served_roi = cuisne_in_roi[index_roi]

def print_cuisine(arr):
    for k in arr:
        print(k)
    print('*****************************')

print("Top 10 cusines served in delhi NCR are :")
print_cuisine(max_served_delhi)
print("Top 10 cusines served in Rest of India are :")
print_cuisine(max_served_roi)







```

```python
max_served_delhi
counts_delhi = counts_delhi[index_delhi]

max_served_roi
counts_roi = counts_roi[index_roi]

plt.pie(counts_delhi,labels = max_served_delhi,autopct = "%.1f")
plt.show()

plt.pie(counts_roi,labels = max_served_roi,autopct = "%.1f")
plt.show()

# As the data is more about Delhi NCR than South India so I would be considering the perecentage than counts which 
# Makes pie graph most appropriate as it gives the percentage distribution among the cusines
#Some of the inferences from the data are :
#1.) North Indian food is the most favoured choice for restaurants whether in Delhi or rest of india.
#2.)Chinese food offered more or less in similar way both in Delhi or rest of india.
#3.) Fastfood restaurtants are served in 15.2% restaurants in Delhi-NCR which is quite high 
#compared to 7.1 % restaurants serving it in rest of India.
#4.) Continental food served more in Rest of India (12.9% of restaurants)compared
#to Delhi NCR where only 4.5% restaurants serve it.
#5.)Italian & Cafe are one of the top choice of Rest of India restaurants which is not that prevalent in Delhi NCR.
#6.)Mughlai is one of the top choice of Delhi NCR restaurants
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


#Reading the dataset and taking data from India and replacing few ncr places to make it delhi-NCR
df2 = pd.read_csv('Dataset/zomato.csv',encoding='latin-1')
df2 = df2[df2['Country Code'] == 1]
rate_count = df2['Aggregate rating'].value_counts()
votes = df2['Votes']
cuisine = df2['Cuisines']
avg_cost=df2['Average Cost for two']
num_cuisine = []
for row in cuisine:
    num = len(row.split(','))
    num_cuisine.append(num)



```

```python

rating = df2['Aggregate rating']
plt.scatter(votes,rating)
plt.xlabel('No of votes')
plt.ylabel('Rating')
plt.show()

#This graph is bit complex to understand , so I have divided it into two parts 
#1.) No of votes below 2000 : Here Rating is very unrelated to votes as it is taking any random values between 0 to 5
#2.) for the second part of graph where votes are greater than 2000 , rating is not going below 3.5 , from
#this we can say that any restaurant having votes more than 2000 will be having atleast of 3.5 rating leaving 1 or 2 outliers.

```

```python
plt.scatter(avg_cost,rating)
plt.xlabel('Average Cost')
plt.ylabel('Rating')
plt.show()

#Here we can see that till the Average Cost is 3000 the Ratings don't have any kind of relation , but if we see 
#the second part of the graph (avg cost > 3000) we can see that minimum rating is 3
```

```python
plt.scatter(num_cuisine,rating)
plt.xlabel('No of Cuisines')
plt.ylabel('Rating')
plt.show()
#Nor do we see any particular relation between the No of cuisines and Rating
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



df3 = pd.read_csv('Dataset/zomato.csv',encoding='latin-1')
df3 = df3[df3['Country Code'] == 1]
#locality = df2.Locality.value_counts()



def wavg(group, avg_name, weight_name):
    d = group[avg_name]
    w = group[weight_name]
    try:
        return (d * w).sum() / w.sum()
    except ZeroDivisionError:
        return d.mean()
    
data = df3.groupby('Locality').apply(wavg,'Aggregate rating','Votes')

weighted_rating = np.array(data)
#index = weighted_rating_rest.argsort()[::-1][:10]
#result = weighted_rating_rest[index]
result = data.sort_values(ascending=False)[:10]
print(result)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df7 = pd.read_csv('Dataset/zomato.csv',encoding='latin-1')
df7 = df7[df7['Country Code'] == 216]

Usa_cuisine = np.array(df7['Cuisines'])
US_Cuis=[]

for row in Usa_cuisine:
    row = str(row)
    arr = row.split(',')
    for j in arr:
        US_Cuis.append(j.strip())
    
Cuisine_name,Cusine_count = np.unique(US_Cuis,return_counts = True)
index = Cusine_count.argsort()[::-1][:10]
res_Cusine_count = Cusine_count[index]
res_Cuisine_name = Cuisine_name[index]

plt.pie(res_Cusine_count,labels= res_Cuisine_name,autopct = "%.1f")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df8 = pd.read_csv('Dataset/zomato.csv',encoding='latin-1')
df8 = df8[df8['Country Code'] == 1]

def wavg(group, avg_name, weight_name):
    d = group[avg_name]
    w = group[weight_name]
    try:
        return (d * w).sum() / w.sum()
    except ZeroDivisionError:
        return d.mean()
    
data5 = df8.groupby('City').apply(wavg,'Aggregate rating','Votes')

data6 = df8.groupby('City').sum()


restaurant_weight = np.array(data5)
City = data5.index
City2 = data6.index
no_rest = data6['Country Code']



plt.scatter(City,no_rest,s =restaurant_weight)
plt.xticks(rotation = 90)
plt.show()

```
