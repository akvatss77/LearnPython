## Cutomizing the graphs

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,5,0.1)
y = x ** 3
plt.plot(x,y,color = 'black',marker = 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,5,0.1)
y = x ** 3
plt.plot(x,y,color = 'black',linewidth = 5)
plt.show()
```

## Giving labels and titles

**pylot.ylablel('shown at y -axis')**<br>
**pylot.xlablel('shown at x -axis')**<br>
**pylot.title('plotting cube function')**<br>
**pylot.legend() shows graph label**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2
plt.plot(x,y,color = 'blue',linewidth = 2,label = 'x^3')
plt.plot(x,y2,color = 'green',linewidth = 2,label = 'x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() 
plt.show()
```

**We can change the starting and ending of axis using pylot.axis() function which takes 4 args**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2
plt.plot(x,y,color = 'blue',linewidth = 2,label = 'x^3')
plt.plot(x,y2,color = 'green',linewidth = 2,label = 'x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend() 
plt.axis([0,10,0,200])
plt.show()
```

# plt.grid() function : this brings the grid on the graph

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2
plt.plot(x,y,color = 'blue',linewidth = 2,label = 'x^3')
plt.plot(x,y2,color = 'green',linewidth = 2,label = 'x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Matlibplot demo')
plt.legend() 
plt.grid()
plt.show()
```

**plt.text() : this function can be used to write some text on graph at a particular coordinate**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,5,0.1)
y = x ** 3
y2 = x ** 2
plt.plot(x,y,color = 'blue',linewidth = 2,label = 'x^3')
plt.plot(x,y2,color = 'green',linewidth = 2,label = 'x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Matlibplot demo')
plt.legend() 
plt.grid()
plt.text(2,80,'text',fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x1 =[0,1,2,3,4,5,6,7,9,10]
y1=[1,2,5,10,17,26,37,50,82,101]
x = np.arange(0,40,2)
y = x.copy()

plt.plot(x1,y1,color='black',label = 'One')
plt.plot(x,y,color='red',label = 'Two')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
