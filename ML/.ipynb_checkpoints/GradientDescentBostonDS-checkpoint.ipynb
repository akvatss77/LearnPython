```python
import numpy as np
import pandas as pd

#Reading the training data in an np array
train_data = np.genfromtxt('Dataset/0000000000002417_training_boston_x_y_train.csv',delimiter=',')

#Reading the testing data in an np array
test_data = np.genfromtxt('Dataset/0000000000002417_test_boston_x_test.csv',delimiter=',')

#Checking the shape of both the arrays
print(test_data.shape)
print(train_data.shape)
```

```python
#Capture the no of features
n = train_data[0].size

#leaving the last column which is value of y
X_train = train_data[:,0:n-1]

#Slicing Y train
Y_train = train_data[:,n-1]

#Creating the additional column with all values as 1 for the intercept c 
add_on = np.ones(len(train_data))

#concatenating it to the X_train
X_train = np.column_stack((X_train, add_on))
```

```python

#cost function to calculate the mean squared error
def cost(X,Y,m):
    total_cost = 0
    #Adding the cost for each of the datapoints(rows)
    for i in range(len(X)):
        x_row = X[i,]
        y = Y[i]
        total_cost += (1/len(X))*((y - (m*x_row).sum())** 2)
    return total_cost

def step_gradient(X,Y,learning_rate,m):
    m_slope = np.zeros(len(m))
    N = len(X)
    for i in range(N):
        x_row = X[i,]
        y = Y[i]
        for j in range(len(m)):
            m_slope[j] += (-2/N)* (y - (m * x_row).sum())* x_row[j]
    for k in range(len(m)):
        m[k] = m[k] - (learning_rate * m_slope[k])
    return m

def gd(X,Y,iteration,learning_rate):
    m = np.zeros(14)
    for i in range(iteration):
        m = step_gradient(X,Y,learning_rate,m)
        print(i,"cost :",cost(X,Y,m))
        
    return m

def run():
    iteration = 10000
    learning_rate = 0.1
    m = gd(X_train,Y_train,iteration,learning_rate)
    print(m)
    return m
    
```

```python
m = run()
```

```python
add_on2 = np.ones(len(test_data))

X_test = np.column_stack((test_data, add_on2))

def predict(X,m):
    Y = np.empty(len(X))
    for k in range(len(X)):
        Y[k] = (m * X[k,]).sum()
    return Y
```

```python
Y = predict(X_test,m)
```

```python
def run_scaled():
    iteration = 10000
    learning_rate = 0.1
    m = gd(X_train_scaled,Y_train,iteration,learning_rate)
    print(m)
    return m
```

```python
m2 = run_scaled()
```
