# Pie graphs

```python
import matplotlib.pyplot as plt
```

```python
sizes = [3,4,6,2]
plt.pie(sizes)
plt.show()
```

## axis command -- we are trying to make the graph uniformly shaped

```python
sizes = [3,4,6,2]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

## changing colors for pie chart

```python
sizes = [3,4,6,2]
colors = ['blue','black','pink','green']
plt.pie(sizes,colors = colors)
plt.axis("equal")
plt.show()
```

## Adding labels to the pie chart

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','black','pink','green']
plt.pie(sizes,colors = colors,labels= labels)
plt.axis("equal")
plt.show()
```

## giving the title

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','black','pink','green']
plt.pie(sizes,colors = colors,labels= labels)
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

## Showing percentages

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','gray','pink','green']
plt.pie(sizes,colors = colors,labels= labels,autopct = "%.1f")# 1f means only one number after decimal
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','gray','pink','green']
plt.pie(sizes,colors = colors,labels= labels,autopct = "%.2f%%")# to add two numbers after decimal and percentage sign
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

## Explode property

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
explode = [0.1,0,0,0]
colors = ['blue','gray','pink','green']
plt.pie(sizes,colors = colors,labels= labels,autopct = "%.2f%%",explode = explode)# to add two numbers after decimal and percentage sign
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

**if we give 1 in explode it will completely come out**

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
explode = [0.1,.2,0,0]
colors = ['blue','gray','pink','green']
plt.pie(sizes,colors = colors,labels= labels,autopct = "%.2f%%",explode = explode)# to add two numbers after decimal and percentage sign
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

## making the graph clockwise

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
explode = [0.1,.2,0,0]
colors = ['blue','gray','pink','green']
plt.pie(sizes,colors = colors,labels= labels,autopct = "%.2f%%",explode = explode,counterclock = False)# to add two numbers after decimal and percentage sign
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

## Start angle can be different

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
explode = [0.1,.2,0,0]
colors = ['blue','gray','pink','green']
plt.pie(sizes,colors = colors,labels= labels,autopct = "%.2f%%",explode = explode,counterclock = False,startangle = 60)# to add two numbers after decimal and percentage sign
plt.title("Split among classes")
plt.axis("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]

plt.pie(sold,labels = company,autopct = "%.1f")
plt.show()


np_arr = np.array(sold)
tota_sum = np_arr.sum()
for i in range(len(company)):
    perc = (sold[i]/tota_sum) * 100
    print(company[i],format(perc,'.1f'))
```
