```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('Dataset/amazon_jobs_dataset.csv')

df["Year"] = df.Posting_date.str.split(",",n = 1,expand=True)[1]
k = df["Year"].value_counts()[::-1]

years = k.keys()
n_yrs = np.array(years)
jobs = k.values
n_jobs =np.array(jobs)
k_ind = n_yrs.argsort()

n_jobs =n_jobs[k_ind]
n_yrs = n_yrs[k_ind]

plt.plot(n_yrs,n_jobs)
plt.show()
for i in range(len(years)):
    print(int(n_yrs[i]),int(n_jobs[i]))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Dataset/amazon_jobs_dataset.csv')
temp = df.Posting_date.str.split(",",n = 1,expand=True)[0]
month = temp.str.split(" ",n = 1,expand = True)[0]
freq = month.value_counts()
x = freq.index
y = freq
plt.bar(x,y)
plt.xticks(rotation = 40)
plt.show()
for i in range(len(x)):
    print(x[i],y[i])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Dataset/amazon_jobs_dataset.csv')
df["Country"] = df.location.str.split(",",expand= True)[0]
df = df[df["Country"] == "IN"]
df["City"] = df.location.str.split(",",expand=True)[2]
freq = df["City"].value_counts()
freq = freq.sort_values(ascending=False)
city = freq.index
jobs = freq
total = freq.sum()

plt.pie(jobs,labels=city,autopct="%.2f%%")
plt.axis("equal")
plt.show()

for i in range(len(freq)):
    perc = (jobs[i]/total)*100
    print(city[i].strip(),format(perc,".2f"))
```
