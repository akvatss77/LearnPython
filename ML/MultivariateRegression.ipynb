```python
import numpy as np

from sklearn import datasets

boston = datasets.load_boston()
```

```python
X = boston.data
Y = boston.target
```

```python
import pandas as pd

df= pd.DataFrame(X)
df.columns = boston.feature_names
```

```python
#now here we need to add all the features
#lets start with one : 'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX','PTRATIO', 'B', 'LSTAT'
#df['CRIM_CRIM'] = df['CRIM'] ** 2
#df['CRIM_ZN'] = df['CRIM'] * df['ZN']
```

```python
cols = boston.feature_names
len(cols)

for i in range(len(cols)):
    k = cols[i] + '_' + cols[i]
    df[k] = df[cols[i]] ** 2
    for j in range(i+1,len(cols)):
        w = cols[i] + '_' + cols[j]
        df[w] = df[cols[i]] * df[cols[j]]
    
```

```python
X2 = df.values
```

```python
from sklearn import model_selection
X1_train,X1_test,Y1_train,Y1_test = model_selection.train_test_split(X,Y,random_state = 0)
X2_train,X2_test,Y2_train,Y2_test = model_selection.train_test_split(X2,Y,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression

alg1 = LinearRegression()
alg1.fit(X1_train,Y1_train)
alg2 = LinearRegression()
alg2.fit(X2_train,Y2_train)
```

```python
score1_train = alg1.score(X1_train,Y1_train)
score1_test = alg1.score(X1_test,Y1_test)
print('score1_train',score1_train)
print('score1_test',score1_test)


score2_train = alg2.score(X2_train,Y2_train)
score2_test = alg2.score(X2_test,Y2_test)
print('score2_train',score2_train)
print('score2_test',score2_test)
```
